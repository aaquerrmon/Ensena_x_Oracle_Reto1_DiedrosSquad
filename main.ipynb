```python
target_size = 256
batch_size = 64
nombre_carpeta = "best_model"
epochs = 200
```

```python
tf.keras.backend.clear_session() #Borra las variables que se hayan podido crear de otros modelos(as evitamos consumir a la tonto) 
# Segundo modelo
num = 16

model = tf.keras.models.Sequential([
    
 tf.keras.layers.Conv2D(num, (3,3), activation='relu', input_shape=(256, 256, 3)),
 tf.keras.layers.Conv2D(num, (3,3), activation='relu'),
 tf.keras.layers.MaxPooling2D(2, 2),
 
 
 tf.keras.layers.Conv2D(num * 2, (3,3), activation='relu'),
 tf.keras.layers.Conv2D(num * 2, (3,3), activation='relu'),
 tf.keras.layers.MaxPooling2D(2,2),


 tf.keras.layers.Conv2D(num * 4, (3,3), activation='relu'),
 tf.keras.layers.Conv2D(num * 4, (3,3), activation='relu'),
 tf.keras.layers.Conv2D(num * 4, (3,3), activation='relu'),
 tf.keras.layers.MaxPooling2D(2,2),

 tf.keras.layers.Conv2D(num * 8, (3,3), activation='relu'),
 tf.keras.layers.Conv2D(num * 8, (3,3), activation='relu'),
 tf.keras.layers.Conv2D(num * 8, (3,3), activation='relu'),
 tf.keras.layers.MaxPooling2D(2,2),
 
 tf.keras.layers.Flatten(),
 tf.keras.layers.Dense(512, activation='relu'),
 tf.keras.layers.Dense(512, activation='relu'),


 tf.keras.layers.Dense(nClases, activation='softmax')
])##Arquitectura de nuestro modelo

model.summary()
```

```python
model, history = train_model(model, epochs, target_size, batch_size, nombre_carpeta)##Entrenamiento
```
